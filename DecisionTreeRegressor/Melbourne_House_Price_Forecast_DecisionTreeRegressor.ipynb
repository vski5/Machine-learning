```python
import pandas as pd
```

```python
Melbourne_housing_FULL_path = "/home/ubuntu/anaconda/date/moerben/Melbourne_housing_FULL.csv"
Melbourne_housing_FULL = pd.read_csv(Melbourne_housing_FULL_path)

display(Melbourne_housing_FULL)
```

```python
Melbourne_housing_FULL.columns
```

```python
Melbourne_housing_FULL = Melbourne_housing_FULL.dropna(axis= 0)
```

```python
y = Melbourne_housing_FULL.Price
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = Melbourne_housing_FULL[melbourne_features]
```

```python
X.describe()
```

```python
X.head(10)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("\n")
print("The predictions are")
print(melbourne_model.predict(X))
```
