```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

用一个需要算方差的模型作为示例

```python
iowa_file_path = "/home/ubuntu/anaconda/date/home-data-for-ml-course/train.csv"
home_data = pd.read_csv(iowa_file_path)

y = home_data.SalePrice
feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[feature_columns]

from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)

```

max_leaf_nodes:最大叶结点 
控制树的深度,允许模型做的叶子越多，就会从欠拟合区域移到过拟合区域。
可以使用一个 实用函数 get_mae() 帮助比较不同的max_leaf_nodes值的MAE得分。
实用函数 get_mae()会先按规定深度和规定随机数预设一个模型。

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    #规定深度和规定随机数预设一个模型
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    #用训练集训练模型
    model.fit(train_X, train_y)
    #模型使用预测集的X预测结果
    preds_val = model.predict(val_X)
    #计算预测值和实际值平均方差
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

用不同的max_leaf_nodes的值比较MAE

```python
candidate_max_leaf_nodes = [x for x in range(999999) if x%5==0 and x>200]
#print(candidate_max_leaf_nodes)
```

```python
maes = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}
```

```python
best_tree_size = min(maes, key=maes.get)

print("OK")
```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=0)

final_model.fit(train_X,train_y)

print("OK")
```

```python
predict_y = final_model.predict(X)
print("OK")
```

```python
final_mae = mean_absolute_error(predict_y, y)
print("OK")
```
