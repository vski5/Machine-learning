```python
import pandas as pd
```

    创造数据 
    一个DataFrame是一个表
    {'Yes': [50, 21], 'No': [131, 2]}
    格式如同一个字典。

```python
pd.DataFrame( {'Yes': [50, 21], 'No': [131, 2]} )
```

```python
pd.DataFrame( {"竖列名称":[1,2,3,"对应的值也可以是字符串"]} )
```

    Series系列是一个数据值的序列。
    如果DataFrame是一个表，那么Series就是一个列表,只有一列无名称的表，
    用索引参数给系列分配行标签，一个Series没有列名，它只有一个整体名称。
    事实上，你可以只用一个列表来创建一个系列。

```python
pd.Series([1, 2, 3, 4, 5])
```

index是个常见的加索引的关键词，不止Series,在DataFrame也能用

```python
pd.Series([30, 35, 40], index=['标签名1', '标签名2', '标签名3'], name='Series的名称')
```

    pd.read_csv()可以指定30多个参数
    CSV文件有一个内置的索引，而pandas不自动接收。
    为了让pandas使用该列作为索引（而不是从头开始创建一个新的索引），我们可以指定一个index_col。
    index_col=0就是把原来的第零列作为索引列。

```python
test_read_index = pd.read_csv("/home/ubuntu/anaconda/date/home-data-for-ml-course/train.csv",index_col=0)
test_read_index.head(6)
#在此处，id就是csv自带的索引
```

```python
#输出数据
test_read_index.to_csv("some_name.csv")
```
