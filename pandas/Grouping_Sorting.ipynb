```python
import pandas as pd
reviews = pd.read_csv("/home/ubuntu/anaconda/data/winemag-data_first150k.csv", index_col=0)
# 设置选项展示的最大行数为5
pd.set_option("display.max_rows", 10)
```

```python
# value_counts()查频率
# 查reviews表中points这一列（groupby('points')），查不同points的记数（也就是频率）points.count()
reviews.groupby('points').points.count()
```

```python
# 查reviews表中points这一列（groupby('points')），
# 查上文提炼出来的points这一列中，不同price的最小价格 price.min()
reviews.groupby('points').points.count()
```

```python
# 表格.apply(表格中每个内容会遭受的处理)
reviews.groupby('winery').apply(lambda df: df.description.iloc[0])
```

```python
# 找出['country', 'province']中loc[df.points.idxmax()] 评分最高的一行
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# 用agg()函数找到 三个特征值。
reviews.groupby(['country']).price.agg([len, min, max])
```

```python
#  groupby()会产生多索引。多索引与普通索引的不同之处在于，它有多个层次。

countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

    最常使用的多索引方法是将其转换为普通索引的方法，即reset_index()方法。
    将普通索引改成按值排列sort_values()

```python
countries_reviewed2 = countries_reviewed.reset_index()
countries_reviewed2
```

```python
#将普通索引改成按值排列sort_values()
#升序排列
countries_reviewed2.sort_values(by='len')
```

```python
# 降序排列
countries_reviewed2.sort_values(by='len', ascending=False)
```

```python
# 在用reset_index()排列成普通索引的条件下，用sort_index()使之按索引排列
countries_reviewed2.sort_index()
```

```python
# 按多个列排序
countries_reviewed2.sort_values(by=['country', 'len'])
```

```python
# 找出 prices一列中points最大的，如何按prices的值排列
best_rating_per_price = reviews.groupby('price')['points'].max().sort_index()
best_rating_per_price
```

```python
# 创建一个DataFrame，其索引是数据集中的品种类别，其值是其最小和最大值。
price_extremes = reviews.groupby('variety').price.agg([min, max])
price_extremes 
```

```python
# 创建一个变量 sorted_varieties，其中包含前一个问题中的数据框架的副本，
# 根据最低价格降序排序，然后根据最高价格排序。
sorted_varieties = price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties
```

```python
# 创建一个系列，其索引是{国家，品种}对的多重索引。
#例如，在美国生产的黑皮诺应该映射到{"美国"，"黑皮诺"}。
#根据酒的数量，将系列中的值按降序排列。
country_variety_counts = reviews.groupby(['country','variety']).size().sort_values(ascending=False)
country_variety_counts
```
