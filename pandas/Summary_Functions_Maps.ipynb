# 先讲lambda函数

```python
#用作筛选x的值 以 成立一个list
list(filter(lambda x : x % 2 == 1, range(1, 20)))
# 此处未知数x的值代入range(1, 20) , 1到20
```

```python
# 用作判断 
(lambda x: x > 2)(3) 
# 此处未知数x的值代入3  => True
```

```python
# 类似def 
(lambda x, y: x ** 2 + y ** 2)(2, 1)
# 此处未知数 x, y 的值代入(2, 1)  => 5
```

# maps，就是映射

```python
review_points_mean = reviews.points.mean()
# 建立points这个数据集的map，所有未知数p都会减去 review_points_mean
reviews.points.map(lambda p: p - review_points_mean)
# 此处的lambda配合map，有点像js里的map配合箭头函数，创造一个新的数据，由reviews.points代入p，并执行lambda函数
```

```python
centered_price = reviews.price - reviews.price.mean()
centered_price
```

    通过调用apply()在每一行上调用一个自定义方法（此处为remean_points()）来转换整个DataFrame。
    map()和apply()分别返回新的、经过转换的Series和DataFrames。它们不会修改它们所调用的原始数据。

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row
# axis用于选择行列，axis=1 (or axis='columns')是一整列上的每行数据都用函数remean_points处理了
reviews.apply(remean_points, axis='columns')
```
